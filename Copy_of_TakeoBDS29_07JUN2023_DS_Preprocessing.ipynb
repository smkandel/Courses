```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Data.csv')
df
```

```python
df_data3 = pd.read_csv('/content/Data3.csv')
df_data3
```

```python
X = df.iloc[:,:-1].values # independent
y = df.iloc[:,-1].values # dependent
```

```python
X1 = df_data3.iloc[:,:-1].values # independent
y1 = df_data3.iloc[:,-1].values # dependent
```

```python
# Missing value

from sklearn.impute import SimpleImputer
# sklearn.impute.SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

imputer.fit(X[:,1:3])
X[:,1:3]= imputer.transform(X[:,1:3])
```

```python
# Missing value

from sklearn.impute import SimpleImputer
# sklearn.impute.SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

imputer.fit(X1[:,3:4])
X1[:,3:4]= imputer.transform(X1[:,3:4])
print(X1)
```

```python
# Missing value

from sklearn.impute import SimpleImputer
# sklearn.impute.SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value= 'India')

imputer.fit(X1[:,4:5])
X1[:,4:5]= imputer.transform(X1[:,4:5])
print(X1)
```

```python
from sklearn.compose import ColumnTransformer # sklearn.compose.ColumnTransformer
from sklearn.preprocessing import OneHotEncoder # sklearn.preprocessing.OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [2,4])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
# sklearn.model_selection.train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train, X_test, y_train, y_test 
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,4:] = sc.fit_transform(X_train[:,4:])
X_test[:,4:] = sc.transform(X_test[:,4:])
```
